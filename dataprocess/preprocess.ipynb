```python
def df_to_lists(df,col):
    df_data_list = df[col].values.tolist()
    df_data_dict = dict(zip(*np.unique(df_data_list,return_counts=True)))

    df_values = list(df_data_dict.values())

    df_keys = list(df_data_dict.keys())
    return df_keys, df_values

ada_acc_keys, ada_acc_values = df_to_lists(playground_data,'ada_accessible')
safety_surface_keys, safety_surface_values = df_to_lists(playground_data,'safety_surface_type')
name_keys, name_values = df_to_lists(playground_data,'name')
manufacturer_keys, manufacturer_values = df_to_lists(playground_data,'manufacturer')
equipment_type_keys, equipment_type_values = df_to_lists(playground_data,'equipment_type')

print('ada_acc')
print(ada_acc_keys,ada_acc_values)

print('safety_surface')
print(safety_surface_keys, safety_surface_values)

print('name_keys, name_values')
print(name_keys, name_values)

print('manufacturer_keys, manufacturer_values')
print(manufacturer_keys, manufacturer_values)

print('equipment_type_keys, equipment_type_values')
print(equipment_type_keys, equipment_type_values)


```
